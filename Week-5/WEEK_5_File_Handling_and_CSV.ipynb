# PART 1: Opening and Reading a File


```python

file = open("week5_student_scores.csv", "r")

for line in file:
    print(line)

file.close()
```


Explanation:

- open(filename, "r") opens file in read mode
- for loop reads each line
- close() releases system resources



# PART 2: Extracting Specific Columns


```python

file = open("week5_student_scores.csv", "r")

next(file)  # Skip header row

for line in file:
    data = line.strip().split(",")
    
    student_id = data[0]
    name = data[1]
    math_score = int(data[2])
    
    print(student_id, name, math_score)

file.close()
```


# PART 3: Compute Average Math Score


```python

file = open("week5_student_scores.csv", "r")

next(file)

total = 0
count = 0

for line in file:
    data = line.strip().split(",")
    math_score = int(data[2])
    
    total += math_score
    count += 1

average = total / count

print("Average Math Score:", average)

file.close()
```


# PART 4: Store Scores in a List


```python

scores = []

file = open("week5_student_scores.csv", "r")
next(file)

for line in file:
    data = line.strip().split(",")
    scores.append(int(data[2]))

file.close()

print("Math Scores:", scores)
```


# PART 5: Find Highest and Lowest Scores


```python

highest = scores[0]
lowest = scores[0]

for score in scores:
    if score > highest:
        highest = score
        
    if score < lowest:
        lowest = score

print("Highest Math Score:", highest)
print("Lowest Math Score:", lowest)
```
