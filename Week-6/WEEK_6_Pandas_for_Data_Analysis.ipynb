# PART 2: Creating a DataFrame

A DataFrame is a table (rows and columns).


```python

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Math": [65, 80, 45, 90, 75],
    "English": [70, 85, 50, 88, 80],
    "Science": [72, 78, 48, 92, 79]
}

df = pd.DataFrame(data)
df
```


# PART 3: Exploring Data


```python

df.head()
```

```python

df.info()
```

```python

df.describe()
```


# PART 4: Selecting Columns


```python

df["Math"]
```

```python

df[["Math", "English"]]
```


# PART 5: Computing Averages


```python

df["Math"].mean()
```

```python

df[["Math", "English", "Science"]].mean()
```


# PART 6: Finding Maximum Values


```python

df["Math"].max()
```

```python

df.max()
```


# PART 7: Filtering Data

Example: Students who scored above 70 in Math


```python

df[df["Math"] > 70]
```


# PART 8: Creating a New Column (Average Score)


```python

df["Average"] = df[["Math", "English", "Science"]].mean(axis=1)
df
```


# PART 9: Creating Pass/Fail Column


```python

df["Result"] = df["Average"].apply(lambda x: "Pass" if x >= 50 else "Fail")
df
```
