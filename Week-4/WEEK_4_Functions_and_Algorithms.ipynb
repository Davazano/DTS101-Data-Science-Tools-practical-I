# PART 1: Basic Function


```

def greet():
    print("Welcome to Data Science")

greet()
```


# PART 2: Function with Parameter


```

def greet_student(name):
    print("Hello", name)

greet_student("David")
greet_student("Mary")
```


# PART 3: Function to Compute Average


```

def compute_average(scores):
    total = 0
    
    for score in scores:
        total = total + score
    
    average = total / len(scores)
    
    return average


scores = [65, 70, 80, 90, 75]

avg = compute_average(scores)

print("Average score:", avg)
```


# PART 4: Function to Determine Pass or Fail


```

def check_pass_fail(score):
    
    if score >= 50:
        return "Pass"
    else:
        return "Fail"


print(check_pass_fail(75))
print(check_pass_fail(40))
```


# PART 5: Apply Function to Multiple Students


```

scores = [45, 67, 80, 30, 90]

for score in scores:
    result = check_pass_fail(score)
    print(score, result)
```


# PART 6: Algorithm Explanation

Algorithm steps to compute average:

1. Start
2. Set total = 0
3. Loop through scores
4. Add scores to total
5. Divide by number of scores
6. Display result
7. End
